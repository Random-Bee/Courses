```python
import numpy as np
import matplotlib.pyplot as plt
import subprocess
import time
```

```python
input_file = "inp-params.txt"
```

```python

outputs = np.zeros((2, 6, 5))

for i in range (1,7):
    
    print(i)
    
    new_content = f'{2**i} 15 1 2'

    print(new_content)
    
    with open(input_file, "w") as f:
        f.write(new_content)
    

    cmd = ["./1.out", input_file]

    for j in range (7):
        process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

        # Wait for the process to finish
        stdout, stderr = process.communicate()

        # Check the return code to see if the execution was successful
        if process.returncode == 0:
            print(f"Execution of '{' '.join(cmd)}' was successful.")
            print("Standard Output:")
            print(stdout)
            arr = stdout.split('\n')
            arr = [it.split(' ') for it in arr]
            arr = arr[1][-1]
            if j<2:
                continue
            outputs[0][i-1][j-2] = float(arr)
        else:
            print(f"Error: Execution of '{' '.join(cmd)}' failed.")
            print("Standard Error:")
            print(stderr)

    cmd = ["./2.out", input_file]

    for j in range (7):
        process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

        # Wait for the process to finish
        stdout, stderr = process.communicate()

        # Check the return code to see if the execution was successful
        if process.returncode == 0:
            print(f"Execution of '{' '.join(cmd)}' was successful.")
            print("Standard Output:")
            print(stdout)
            arr = stdout.split('\n')
            arr = [it.split(' ') for it in arr]
            arr = arr[1][-1]
            if j<2:
                continue
            outputs[1][i-1][j-2] = float(arr)
        else:
            print(f"Error: Execution of '{' '.join(cmd)}' failed.")
            print("Standard Error:")
            print(stderr)

print(outputs)

results = np.zeros((2, 6))

for i in range (0,2):
    for j in range (0,6):
        results[i][j] = np.mean(outputs[i][j])

print(results)
```

```python
y = [2, 4, 8, 16, 32, 64]
results[1][0] = 10891.121
plt.plot(y, results[0], label="Throughput for CLH Lock")
plt.plot(y, results[1], label="Throughput for MCS Lock")
plt.scatter(y, results[0])
plt.scatter(y, results[1])
plt.ylabel("Throughput")
plt.xlabel("No. of Threads")
plt.legend()
plt.show()
```
