```python
def getEdges(n):
    for i in range(n):
        print(i+1, end=" ")
        for j in range(n):
            # rnd = np.random.rand()
            # if i!=j and rnd < 0.5:
            if i!=j:
                print(j+1, end=" ")
        print()

getEdges(15)
```

```python
x = [10, 11, 12, 13, 14, 15]

y1 = [10, 11, 12, 13, 14, 15]
y2 = [72, 76, 80, 84, 88, 92]

z1 = [
    (7.632+7.906+7.604)/3,
    (8.43273+8.51455+8.67818)/3,
    (9.60167+9.95333+9.58333)/3,
    (10.5108+10.6923+10.7369)/3,
    (11.5214+11.6343+11.48)/3,
    (12.2307+11.9813+12.1587)/3
]
z2 = [124, 128, 132, 136, 140, 144]
```

```python
plt.plot(x, y1, label="VC")
plt.plot(x, z1, label="SK")
plt.xlabel("Number of communicating processes")
plt.ylabel("Average number of entries sent")
plt.title("Number of entries sent")
plt.legend()
plt.show()
```

```python
plt.plot(x, y2, label="VC")
plt.plot(x, z2, label="SK")
plt.xlabel("Number of communicating processes")
plt.ylabel("Max space used (in bytes)")
plt.title("Space used")
plt.legend()
plt.show()
```
