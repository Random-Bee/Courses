```python
import scipy.optimize as opt
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x[0]**2 + x[1]**2
```

```python
x = [1,1]
mn = opt.fmin(f,x,full_output=True)
mn
```

```python
def f1(x):
    return (x[0]-1)**2 + (x[1]-2.5)**2
```

```python
cons = ({'type': 'ineq', 'fun': lambda x: x[0]-2*x[1]+2},
        {'type': 'ineq', 'fun': lambda x: -x[0]-2*x[1]+6},
        {'type': 'ineq', 'fun': lambda x: -x[0]+2*x[1]+2})
bnds = ((0,None), (0,None))
```

```python
res=opt.minimize(f1, (2,0), method='SLSQP', bounds=bnds, constraints=cons)
res
```

```python
def f2(x):
    return (x**3-1)
```

```python
def f2p(x):
    return 3*x**2
```

```python
sol = opt.root_scalar(f2, x0=0.2, fprime=f2p, method='newton')
sol
```

```python
def odes(t,x):
    y=x[0]
    p=x[1]
    dydt=(-0.4*y)+(0.02*p*y)
    dpdt=(0.8*p)-(0.01*p**2)-(0.1*p*y)
    return [dydt, dpdt]
```

```python
t = np.array([0,60])
times=np.linspace(t[0],t[1],600)
x0=[20,20]
```

```python
from scipy.integrate import solve_ivp
```

```python
sol = solve_ivp(odes, [t[0],t[-1]], y0=x0, t_eval=times)
sol
```

```python
t=sol.t
y=sol.y[0]
p=sol.y[1]
```

```python
plt.figure(figsize=(8,8))
plt.plot(t,y,label='y')
plt.plot(t,p,label='p')
plt.legend()
```
